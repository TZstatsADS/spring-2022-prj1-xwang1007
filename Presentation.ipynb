## Part 1 Word Cloud

For the **first question**, how the vocabulary often used by philosophers changed over three periods, I decided to use Word Cloud. Intuitively, the more times a word appears in text resources, the larger and bolder it becomes in the word cloud, and it may be more important.

The following three figures respectively represent word clouds from three periods, namely, ancient, medieval and modern. We can observe some association between word clouds and events or thought during that period.

From the first picture, we can see words such as 'epic', 'indicted', and 'effect' were most important, while 'socrates', 'athenians', and 'prosecuting' were also mentioned a lot. The result makes sense because Socrates is known for his quest for truth, and for Apology of Socrates, which was compiled and written by his student Plato.

The second picture showed that the most frequent words have changed to 'women', 'rights', and 'hope', which indicated that philosophers from the middle age shifted their interest to humans. Some famous sentences, such as I think, therefore I am, nicely supported this change.

The third-word cloud gave us some information that the modern philosophers began to refer to words like 'psychologists', 'sychology', and 'physicists'. With the development of psychology, modern philosophers have also kept up with the times and combined their doctrines with it.


## Part 2 Sentiment Analysis

To start with sentiment analysis, I built a Bi-directional LSTM multi-label sentiment classification model. The model was trained using the semeval2018 dataset. Applying the trained model on cleaned Twitter data for each month, I created a sentiment evolving graph for the three periods.

The flow chart provided information on the Bi-directional LSTM Architecture. I employed the GloVe (Global Vectors for Word Representation) at the embedding layer to generate an embedding matrix for the input texts. Then I made a bi-directional LSTM layer with 128 units each. Finally, I used the Sigmoid activation function to produce multi-label outputs. The model produces a multi-label accuracy of 67% and an F1-micro score of 0.57, which I think is pretty great.

For **question 2**, were ancient philosophers more pessimistic than modern philosophers, the sentiment analysis would be of great help. I defined the sentiment rate (for each sentiment of three periods) to be the ratio of the total number of sentences with that sentiment to all sentences of the period.

The line graph gave us the answer to question 2 to some degree, that is, philosophers from the middle age were the most pessimistic.
They felt more disgusting, fearful, and sad than the other two groups. In addition, ancient philosophers were more pessimistic than modern philosophers.

In order to have a more intuitive look at the positive and negative emotions of philosophers in different periods, on the basis of the line graph, I created a bar graph. As depicted in the bar chart, when I divided the eleven feelings into two groups: (anticipation, joy, love, optimism, and trust) and negative (anger, disgust, fear, pessimism, and sadness), we can see that Philosophers expressed more negative emotions than good emotions during all times, which I think was in line with the sad undertones of philosophy. In general, the total rate of negative emotions is around double that of positive emotions.

## Part 3 LDA

For **question 3**, what were the most popular Philosophical topics in different times? To discover the topics, I used Latent Dirichlet Allocation (LDA) algorithm. First, I used the LDA Mallet model with a range of total topic number values on all phrases in each theme. The model with the highest coherence score was picked as the best model. The best value for the number of subjects is the total topic number. Then, using this ideal topic number, I recreated an LDA Mallet model. Finally, I used PyLDAVis, a visualization tool, to show the topic modeling results in order to have a better and more intuitive grasp of the topic modeling results.

Now I applied the LDA Mallet model and created 2 topics.
